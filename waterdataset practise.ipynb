```python
y_true = ['1','0','0','1','0','1','1','0']
y_pred = ['1','0','1','1','1','1','1','1']
```

```python
print(confusion_matrix(y_true,y_pred))
```

```python
count = 0
for i,j in zip(y_true,y_pred):
    if(i==j):
        count += 1
print(count)
    
```

```python
accuracy = count/len(y_pred)
```
